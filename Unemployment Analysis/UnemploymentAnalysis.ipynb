```python
data.rename(columns={'Region': 'State'}, inplace= 'true')
```

```python
data_clean=data.dropna(how='all')
```

```python
data_clean.isnull().sum()
```

```python
data_clean.columns=['State','Date','Time Period','Estimated Unemployment Rate (%)','Estimated Employed','Estimated Labour Participation Rate (%)','Area']	
data_clean
```

```python
data_format=data_clean.copy()
data_format['Date']=pd.to_datetime(data_clean['Date'],dayfirst='true')
data_format['Month']=data_format['Date'].dt.month
data_format['Year']=data_format['Date'].dt.year
data_format
```

```python
data_format.State.unique()
```

```python
data_format.State.nunique()
```

```python
df_2019=data_format[data_format['Year'] == 2019]
```

```python
df_2019
```

```python
df_2020=data_format[data_format['Year'] == 2020]
```

```python
df_2020
```

```python
df_urban_2019=df_2019[df_2019['Area'] == 'Urban']
df_urban_2019
```

```python
df_rural_2019=df_2019[df_2019['Area'] == 'Rural']
df_rural_2019
```

```python
df_urban_2020=df_2020[df_2020['Area'] == 'Urban']
df_urban_2020
```

```python
dr_gouped_urban_2019=df_urban_2019[['State','Estimated Unemployment Rate (%)']].groupby('State').mean().sort_values('Estimated Unemployment Rate (%)',ascending=False )
dr_gouped_urban_2019.head()
```

```python
dr_gouped_urban_2019.plot.bar()
plt.title('Urban 2019 Highest Rates')
plt.show()
```
